What does a low disparity value mean?

Question 6 options:

Disparity does not tells us about the position of the pixels, but about how similar they are


There are no matching pixels between the two images


# Answer

A low disparity value means that the matching pixels are found near the same position in both images.

In stereo vision, disparity is the difference in horizontal pixel coordinates between the corresponding points in a pair of stereo images. A low disparity value indicates that the corresponding points in the left and right images are close to each other horizontally. # Answer


	
ICP works best if an initial rough estimation of the alignment of 2 point clouds is known. 



ICP (Iterative Closest Point) is a widely used algorithm for registration of two or more point clouds. It aims to find the rigid transformation that best aligns one point cloud to another by minimizing the distance between corresponding points. The algorithm iteratively alternates between finding correspondences between points in the two clouds and estimating the transformation that aligns them.

ICP assumes that the transformation between the two point clouds is rigid, meaning it involves only rotation and translation. Therefore, the correct statement is "ICP provides a rigid transformation between 2 point clouds".

ICP does not necessarily require an initial rough estimation of the alignment of the two point clouds, although it can be beneficial in some cases. ICP is not guaranteed to provide the optimal transformation that best aligns two point clouds, as it can get stuck in local minima. ICP can work well even when the point clouds are close to each other, but it may require a good initialization or some modifications to the algorithm. Finally, ICP provides only a rigid transformation, not an affine transformation.

Consider a color image with dimensions 640x480 pixels. We would like to cluster its pixels making full use of both geometric and color information (including intensity/lightness).

How many clusters will the Mean Shift algorithm consider initially?
(provide the exact number below)

# Answer

The Mean Shift algorithm does not have a fixed number of clusters that it considers initially. Instead, it is an iterative algorithm that starts with an initial set of candidate cluster centers and moves them towards the mode of the data distribution until convergence.

The number of clusters that the Mean Shift algorithm will converge to depends on several factors, such as the initial set of candidate cluster centers, the bandwidth parameter, and the data distribution. It is also possible to control the number of clusters by setting a stopping criterion based on the distance between the candidate cluster centers or the number of iterations.

Therefore, it is not possible to determine the exact number of clusters that the Mean Shift algorithm will consider initially for a given image without additional information about the specific implementation and parameters used.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load data from file
X = np.loadtxt('clusters.txt')

# Fit K-means models for different numbers of clusters
inertia_values = []
for n_clusters in range(1, 11):
    km = KMeans(n_clusters=n_clusters)
    km.fit(X)
    inertia_values.append(km.inertia_)

# Plot inertia values as a function of the number of clusters
plt.plot(range(1, 11), inertia_values, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

PCA can create new features based on the original features, but the number of new features is not necessarily equal to the number of original features or the dimensionality of the space.
