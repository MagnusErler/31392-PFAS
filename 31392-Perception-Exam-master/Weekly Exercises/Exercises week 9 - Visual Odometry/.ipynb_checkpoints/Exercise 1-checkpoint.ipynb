```python
# K = np.array([[7.188560e+02, 0.000000e+00, 6.071928e+02], # camera matrix
#               [0, 7.188560e+02, 1.852157e+02],
#               [0, 0, 1]])

# reference_img = np.load("img_" + str(0) + ".npy")

# for t in range(1, 6):

#     # the image at current time=t
#     curImage = np.load("img_" + str(t) + ".npy")
#     # the 3D landmarks in the world coordinates which have been computed in time=t-1
#     landmark_3D = np.load("landmark_" + str(t-1) + ".npy")
#     # the 2D coordinates of the 3D points in the previous frame at time=t-1
#     reference_2D = np.load("reference_2D_" + str(t-1) + ".npy")
    
#     # the 2D landmarks at the current time = t
#     landmark_3D, reference_2D, tracked_2Dpoints = featureTracking(reference_img, 
#                                                                   curImage, 
#                                                                   reference_2D,
#                                                                   landmark_3D)
    
#     """
#     Using OpenCV, implement PnP using Ransac
#     """
#     #_, rvec, tvec, inliers = ...


#     """
#     Transform the translation and rotation into the world frame
#     """
    
    
#     print(tvec[0], tvec[1], tvec[2], rvec[0], rvec[1], rvec[2])

#     # update for next timestep
#     reference_img = curImage


###################################################################################

K = np.array([[7.188560e+02, 0.000000e+00, 6.071928e+02], # camera matrix
              [0, 7.188560e+02, 1.852157e+02],
              [0, 0, 1]])

reference_img = np.load("img_" + str(0) + ".npy")

R = []
T = []
for t in range(1, 6):

    # the image at current time=t
    curImage = np.load("img_" + str(t) + ".npy")
    # the 3D landmarks in the world coordinates which have been computed in time=t-1
    landmark_3D = np.load("landmark_" + str(t-1) + ".npy")
    # the 2D coordinates of the 3D points in the previous frame at time=t-1
    reference_2D = np.load("reference_2D_" + str(t-1) + ".npy")
    
    # the 2D landmarks at the current time = t
    landmark_3D, reference_2D, tracked_2Dpoints = featureTracking(reference_img, 
                                                                  curImage, 
                                                                  reference_2D,
                                                                  landmark_3D)
    
    """
    Using OpenCV, implement PnP using Ransac
    """
    #_, rvec, tvec, inliers = ...
    _, rvec, tvec, inliers = cv2.solvePnPRansac(landmark_3D,tracked_2Dpoints, K, None)

    """
    Transform the translation and rotation into the world frame
    """
    # what we get from the last step is the rotation and translation vectors to transform the world coordinates into camera coordinates
    rot,_ = cv2.Rodrigues(rvec)
    rot = pinv(rot)
    rvec,_ = cv2.Rodrigues(rot)
    tvec = -np.dot(rot, tvec)
    R.append(rvec)
    T.append(tvec)
    print(tvec[0], tvec[1], tvec[2], rvec[0], rvec[1], rvec[2])

    # update for next timestep
    reference_img = curImage
```

What approximate direction did the camera move in?

```python
# Now we have the pose of the camera, we can reconstruct the movement
from mpl_toolkits.mplot3d import axes3d, Axes3D
fig = plt.figure()
ax = Axes3D(fig)
color = ['r', 'g', 'b']
for r, t in zip(R, T):
    r = cv2.Rodrigues(r)
    ax.scatter(t[0], t[1], t[2], color='k')
    for i in range(3):
        o = r[0][:,i]
        d = o+t.flatten()
        ax.plot([t[0], d[0]], [t[1], d[1]], [t[2], d[2]], color=color[i])
plt.xticks(np.arange(0, 5, step=0.5))
plt.yticks(np.arange(0, 5, step=0.5))
plt.show()
```

```python
# Find current directory
import os
direction_array = []
cwd = os.getcwd()
final_path = ""



K = np.array([[7.188560e+02, 0.000000e+00, 6.071928e+02],  # camera matrix
              [0, 7.188560e+02, 1.852157e+02],
              [0, 0, 1]])

reference_img = np.load("img_0.npy")

for t in range(1, 6):
    print(f"\nProcessing image {t}")

    # the image at current time=t
    curImage = np.load(f"img_{t}.npy")
    # the 3D landmarks in the world coordinates which have been computed in time=t-1
    landmark_3D = np.load(f"landmark_{t-1}.npy")
    # the 2D coordinates of the 3D points in the previous frame at time=t-1
    reference_2D = np.load(f"reference_2D_{t-1}.npy")

    # the 2D landmarks at the current time = t
    landmark_3D, reference_2D, tracked_2Dpoints = featureTracking(reference_img,
                                                                  curImage,
                                                                  reference_2D,
                                                                  landmark_3D)

    """
    Using OpenCV, implement PnP using Ransac
    """
    _, rvec, tvec, inliers = cv2.solvePnPRansac(landmark_3D, tracked_2Dpoints, K, None,
                                                iterationsCount=100,
                                                reprojectionError=0.01,
                                                confidence=0.99,
                                                flags=cv2.SOLVEPNP_P3P)

    # Transform the translation and rotation into the world frame
    R, _ = cv2.Rodrigues(rvec)

    t = -R.T.dot(tvec) 


    # Print the results
    print(f"tvec: {t}")
    print(f"rvec: {rvec}")
    print(f"R: {R}")
    print(f"t: {t}")

    # print(tvec[0], tvec[1], tvec[2], rvec[0], rvec[1], rvec[2])

    # Calculate direction did the camera move in the world frame
    direction = np.dot(R, t)

    # flatten the direction vector to a 1D array
    direction = [dir[0] for dir in direction]
    direction_array.append(direction[:2])
    print(f"direction: {direction}")


    # Visualize the results
    # Subplot with two plots
    fig1=plt.figure(figsize=(15,10))
    ax0 = fig1.add_subplot(311)
    ax0.imshow(curImage)

    ax1=fig1.add_subplot(312,projection='3d', autoscale_on=True)
    # Plot the 3D points
    ax1.scatter(landmark_3D[:, 0], landmark_3D[:, 1], landmark_3D[:, 2], c='r', marker='o')
    ax1.set_xlabel('X Label')
    ax1.set_ylabel('Y Label')
    ax1.set_title('3D Landmarks')

    # Show direction on a plot with large arrow
    #ax.quiver(0, 0, 0, direction[0], direction[1], direction[2], length=0.1)

    # plot 2D points on separate plot
    ax2=fig1.add_subplot(313)
    ax2.scatter(tracked_2Dpoints[:, 0], tracked_2Dpoints[:, 1], c='r', marker='o')
    ax2.set_xlabel('X Label')
    ax2.set_ylabel('Y Label')
    ax2.set_title('2D Landmarks')

    plt.show()

    # update for next timestep
    reference_img = curImage

# scatter plot from direction_array
direction_array = np.array(direction_array)
print(direction_array[:, 0], direction_array[:, 1])
plt.scatter(direction_array[:, 0], direction_array[:, 1])
```
